```python
import yfinance as yf
```

```python
start_date = '2020-01-01'
end_date = '2023-12-31'

acoes = ['PETR4.SA', 'VALE3.SA', 'MGLU3.SA']

dados_brutos = yf.download(acoes, start=start_date, end=end_date, auto_adjust=False)
```

```python
dados_acoes = dados_brutos['Adj Close']
```

```python
retornos_diarios = dados_acoes.pct_change()
retornos_diarios.head()
```

```python
retornos_acumulados = (1 + retornos_diarios).cumprod()
retornos_acumulados.head()
```

```python
volatilidade_anual = retornos_diarios.std() * (252**0.5)
volatilidade_anual
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
retornos_acumulados.plot(title="Retorno acumulado dos ativos")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
pesos = [0.4, 0.4, 0.2]
pesos = pd.Series(pesos, index=retornos_diarios.columns)
retorno_portfolio = (retornos_diarios * pesos).sum(axis=1)
retorono_medio_portfolio = retorno_portfolio.mean() * 252
volatilidade_portfolio_anualizada = retorno_portfolio.std() * (252**0.5)
```

```python
matriz_correlacao = retornos_diarios.corr()
matriz_correlacao
```

```python
import seaborn as sns
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Matriz de correlacao dos ativos")
plt.show()
```

```python
import numpy as np
media_retorno = retorno_portfolio.mean()
std_retorno = retorno_portfolio.std()
dias = 252
simulacoes = 1000

resultados_simulacao = pd.DataFrame()
resultados = []
for i in range(simulacoes):
    retornos_futuros = np.random.normal(loc=media_retorno, scale=std_retorno, size=dias)
    caminho_simulado = (1 + retornos_futuros).cumprod()
    df_temp = pd.DataFrame(caminho_simulado)
    resultados.append(df_temp)

resultados_simulacao = pd.concat(resultados, axis=1)
plt.plot(resultados_simulacao)
plt.title('Simulacao de monte carlo do portfolio')
plt.xlabel('Dias')
plt.ylabel('Retorno Acumulado')
plt.show()
```
